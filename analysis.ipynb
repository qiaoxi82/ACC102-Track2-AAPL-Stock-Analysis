```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from D drive, skip first 2 rows (Price and Ticker headers)
data = pd.read_csv(
    r"D:\AAPL_data.csv",
    header=None,
    index_col=0,
    parse_dates=True,
    skiprows=2  
)

# Rename columns correctly
data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

# Convert all columns to numeric
for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Show clean preview
print("Cleaned data preview (first 5 rows):")
print(data.head())
print("\nData types:")
print(data.dtypes)
```

```python
# Drop rows with all NaN values (removes the extra 'Date' header row)
data = data.dropna(how='all')

# Print updated preview
print("Cleaned data preview (after removing extra header row):")
print(data.head())
```

```python
# Summary statistics for the closing price
print("=== Summary Statistics for AAPL Closing Prices ===")
print(data['Close'].describe())
```

```python
# Set plot style and size
plt.figure(figsize=(14, 7))

# Plot the closing price
plt.plot(data['Close'], label='Closing Price', color='#FF6B6B', linewidth=1.5)

# Add title and labels
plt.title('AAPL Stock Price Trend (2020 - 2025)', fontsize=16, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)

# Save the figure to D drive
plt.savefig(r'D:\AAPL_Closing_Price_Trend.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Calculate daily percentage return
data['Daily_Return'] = data['Close'].pct_change()

# Calculate 20-day and 50-day moving averages
data['MA20'] = data['Close'].rolling(window=20).mean()
data['MA50'] = data['Close'].rolling(window=50).mean()

# Display the first 5 rows with new columns
print("=== Data with Daily Returns and Moving Averages ===")
print(data[['Close', 'Daily_Return', 'MA20', 'MA50']].head())
```

```python
# Set plot style and size
plt.figure(figsize=(14, 7))

# Plot price and moving averages
plt.plot(data['Close'], label='Closing Price', color='#FF6B6B', alpha=0.5, linewidth=2)
plt.plot(data['MA20'], label='20-Day Moving Average', color='#4ECDC4', linewidth=1.5)
plt.plot(data['MA50'], label='50-Day Moving Average', color='#FFE66D', linewidth=1.5)

# Add title and labels
plt.title('AAPL Price with Moving Averages (2020 - 2025)', fontsize=16, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)

# Save the figure to D drive
plt.savefig(r'D:\AAPL_Price_With_MA.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Export the fully cleaned and processed data to D drive
data.to_csv(r'D:\AAPL_Final_Cleaned_Data.csv')
print("✅ Final cleaned dataset saved to D drive as 'AAPL_Final_Cleaned_Data.csv'")
```
